```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def FirstDeriv(f,a,b,N):
    '''
    Input: f - function to take derivatives of
    [a,b] - domain of interest
    N - number of intervals
    Output: df - approximation of first derivative of f
    x - numerical grid
    '''
    
    # Numerical grid
    x = np.linspace(a, b, N+1)
    h = x[1] - x[0]

    df = np.empty(N+1)
    y = f(x)
    
    # Interior Grid points: Three-point centered difference formula for approximate derivative df
    df[1:-1] = (y[2:] - y[:-2])/(2*h)

    # Boundary points: one-sided difference formulas
    df[0] = (-3*y[0] + 4*y[1] - y[2])/(2*h)
    df[-1] = (y[-3] - 4*y[-2] + 3*y[-1])/(2*h)
    
    return df, x
```

```python
# Take derivative of f(x) = exp(x)
f = lambda x: np.exp(x)

a = -1
b = 1
N = 10

# Approximation of the first derivative
df,x = FirstDeriv(f,a,b,N)

# Compare approximation and exact solution
plt.figure()
plt.plot(x,df,'-o',label='df approximation')
plt.plot(x,f(x),label='Exact derivative')
plt.xlabel('x')
plt.ylabel('df/dx')
plt.legend()
plt.show()
```
