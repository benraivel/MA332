## Newton-Ralphson Method

The Newton-Ralphson method is an iterative root finding method. Starting with an initial guess $x_0$ for the root $x_*$, the iteration continues with
$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}, \qquad f'(x_n) \neq 0.
$$
